```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
file_path = 'Resources/zomato.csv'
df= pd.read_csv(file_path,  encoding = "ISO-8859-1")
df.head()
```

```python
df1= df.drop(['Country Code','Address','Locality','Locality Verbose','Longitude','Latitude','Currency'],axis=1)
df1
```

```python
df2= df1.drop(['Restaurant Name','Is delivering now','Switch to order menu','Rating color'], axis= 1)
df2
```

```python
# Copy the Aggregate rating column into new Aggregate rating binary column
df2['Aggregate rating binary'] = df2['Aggregate rating']
df2.head()
```

```python
# Convert 'Aggregate rating binary' column to binary values
df2.loc[df2["Aggregate rating binary"] < 4, "Aggregate rating binary"] = 0
df2.loc[df2["Aggregate rating binary"] >= 4, "Aggregate rating binary"] = 1
df2
```

```python
df2 = df2.astype({'Aggregate rating binary' :int })
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
df2.drop(['Aggregate rating'], axis= 1)
```

### Test train split

```python
y = df2['Aggregate rating binary']
X= df2.drop(columns= 'Aggregate rating binary')
X= pd.get_dummies(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
   y,  random_state=1, stratify=y)
X_train.shape
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
results = pd.DataFrame({
   "Prediction": y_pred,
   "Actual": y_test
}).reset_index(drop=True)
results.head()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
