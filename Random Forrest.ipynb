```python
# Initial imports.
import pandas as pd
from pathlib import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
file_path = 'Resources/zomato.csv'
df= pd.read_csv(file_path,  encoding = "ISO-8859-1")
df.head()
```

```python
df1= df.drop(['Country Code','Address','Locality','Locality Verbose','Longitude','Latitude','Currency'],axis=1)
df1
```

```python
df2= df1.drop(['Restaurant Name','Is delivering now','Switch to order menu','Rating color'], axis= 1)
df2
```

```python
# Copy the Aggregate rating column into new Aggregate rating binary column
df2['Aggregate rating binary'] = df2['Aggregate rating']
df2.head()
```

```python
# Convert 'Aggregate rating binary' column to binary values
df2.loc[df2["Aggregate rating binary"] < 4, "Aggregate rating binary"] = 0
df2.loc[df2["Aggregate rating binary"] >= 4, "Aggregate rating binary"] = 1
df2
```

```python
df2 = df2.astype({'Aggregate rating binary' :int })
```

```python
df2.drop(['Aggregate rating'], axis= 1)
```

```python
df2.dtypes
```

```python
df3= df2.drop(['City', 'Cuisines', 'Aggregate rating','Has Table booking', 'Has Online delivery','Rating text'], axis= 1)
```

```python
df3.dtypes
df3.drop(['Average Cost for two'], axis = 1)
```

```python
X= df3.copy()
X= X.drop("Aggregate rating binary",axis=1)
X.head()
```

```python
y = df3['Aggregate rating binary'].ravel()
y[:5]
```

```python
# Splitting into Train and Test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
df3.dtypes
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# Calculate feature importance in the Random Forest model.
importances = rf_model.feature_importances_
importances
```
